```python
import pandas as pd
import numpy as np
import nltk
import os
import nltk.corpus
```

```python
f2 = open('WAOtext.txt', 'r', encoding="utf8")
```

```python
a = []
for line in f2:
    str = line.split()
    a.append(str)
    print(str)
```

```python
a2 = [x for x in a if x != []]
```

```python
for i in range(len(a2)):
    for n in range(len(a2[i])):
        if (a2[i][n] == 'استان') & (n<len(a2[i])-1):
            print(a2[i][n+1])
        if a2[i][n] == 'نود':
            print(a2[i][n+2])
        if a2[i][:3] == ['تعداد', 'تكنولوژي', 'قطع']:
            print(a2[i][4])
        if (a2[i][n] == 'زمان'):
            print(a2[i])
        if a2[i][n] == 'علت':
            print(a2[i])
```
